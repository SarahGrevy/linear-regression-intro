# Regression models

```python
# Here is a cell with a linear regression on mtcars to get you started
# mpg ~ wt + disp
# modify the formula and paste in subsequent cells to run different models
# you can use forward selection, backwards selection, or your own approach

import statsmodels.formula.api as smf

model = smf.ols("mpg ~ wt + disp", data=mtcars)
results = model.fit()
results.summary()
```

Remember, these aren't just abstract numbers in a vacuum. They are real data about real cars. Each variable has a meaning. Before just dropping it in your regression, as yourself what it means to explain the variance in `mpg` using that variable. Does that **methodological choice** make sense? 

```python
model = smf.ols("qsec ~ hp + carb", data=mtcars)
results = model.fit()
results.summary()
```

Interpretation: Fifty percent of the variance in acceleration is explained by both horsepower and the number of carbs. 

The number of carbs explaining more of the variance than the hp based on the coefficient.

The bigger a car's engine is the slower the acceleration.

The more carbs there is, the slower the car is. 

```python
model = smf.ols("am ~ wt + cyl", data=mtcars)
results = model.fit()
results.summary()
```

Interpretation: Weight and the number of cylinders predict only 45% of the variance among automatic and manual cars.

Based on the model, automatic cars tend to be lighter manual and have more cylinders.

Based on the p values, number of cylinders does not predict the type of car transmission you would get from car dealers. However, weight is highly predictive.

```python
model = smf.ols("mpg ~ hp + wt:qsec + am", data=mtcars)
results = model.fit()
results.summary()
```

Interpretation: Horsepower, car transmission, and the interaction between weight and qsec explain 80% of variance in fuel efficiency.

Based on the model, the relationship between the type of transmission a car has and how much it saves fuel is not statistically significant, meaning, the former does not help predict the latter.

A bigger horsepower is negatively correlated to fuel efficiency. Additional hp decreases fuel efficiency by 0.05 miles per gallon.

Test for multicollinearity:

Presence of such determined if number is greater than 0.7:



We reject the null hypothesis that there is multicollinearity between wt:qsec.
